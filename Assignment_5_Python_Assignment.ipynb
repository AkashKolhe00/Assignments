2. Explain the difference between a for loop and a while loop in Python.

In Python, both for and while loops are used for repeating a block of code multiple times, but they differ in how and when they iterate.

For Loop
Purpose: A for loop is typically used when you know in advance how many times you want to execute a block of code. It iterates over a sequence (like a list, tuple, dictionary, string, or range).
Syntax:
python

for item in sequence:
    # Code block to execute
Example:
python

for i in range(5):
    print(i)
This loop will print numbers from 0 to 4, iterating over the sequence generated by range(5).
While Loop
Purpose: A while loop is used when the number of iterations is not known beforehand and depends on a condition being met. It keeps executing the block of code as long as the specified condition is True.
Syntax:
python

while condition:
    # Code block to execute
Example:
python

i = 0
while i < 5:
    print(i)
    i += 1
This loop will also print numbers from 0 to 4, but the iteration continues as long as i is less than 5.
Key Differences
Control:
A for loop is controlled by the elements in the sequence or iterable, automatically handling the iteration.
A while loop is controlled by a condition that needs to be manually managed within the loop (e.g., by incrementing a variable).
Use Case:
Use a for loop when you have a predetermined sequence to iterate over.
Use a while loop when you need to repeat an action until a condition changes, and the number of iterations isn’t fixed.
Example of When to Use Each
For Loop: Iterating over items in a list:
python
colors = ['red', 'green', 'blue']
for color in colors:
    print(color)
While Loop: Running until a condition is met:
python
password = ''
while password != 'secret':
    password = input('Enter the password: ')

4. How do you iterate through a list using a for loop in Python?

Basic Iteration
You can directly loop over each element in the list:

python

my_list = ['apple', 'banana', 'cherry']

for item in my_list:
    print(item)
Explanation
my_list: This is the list you want to iterate over.
for item in my_list:: The for loop iterates over each element in my_list. On each iteration, the variable item is assigned the value of the current element.
print(item): The code block inside the loop will execute for each element, in this case, printing the element.
Output
This code will print:


apple
banana
cherry
Iterating with Index
If you need the index of each element while iterating, you can use the enumerate() function:

python

my_list = ['apple', 'banana', 'cherry']

for index, item in enumerate(my_list):
    print(index, item)
Explanation
enumerate(my_list): This function returns both the index and the value of each element in the list.
for index, item in enumerate(my_list):: On each iteration, index is assigned the current index, and item is assigned the current element.
Output
This code will print:


0 apple
1 banana
2 cherry
This way, you can iterate through a list using a for loop in Python, either by directly accessing the elements or by also getting their indices.

23. Write a Python program to find the LCM (Least Common Multiple) of two numbers using a for loop.

```python
def find_lcm(x, y):
    
    max_num = max(x, y)

    
    for lcm in range(max_num, x * y + 1, max_num):
        if lcm % x == 0 and lcm % y == 0:
            return lcm

find_lcm(24, 15)
```

3. How does the `map()` function differ from a list comprehension in Python, and when would you choose one over the other?

The map() function and list comprehensions in Python are both tools used to apply a function to each item in an iterable, but they have some differences in syntax, flexibility, and performance. Here’s a breakdown of the differences and when you might choose one over the other:

map() Function
Syntax: map(function, iterable)
Purpose: Applies a given function to each item in an iterable (like a list, tuple, etc.) and returns a map object, which is an iterator. This means that the map object must be converted to a list or another iterable type if you want to work with it directly.
When to Use:
When you already have a function defined (e.g., a built-in function or a custom one) and want to apply it directly.
When working with multiple iterables: map() can take multiple iterables as arguments, applying the function to corresponding elements.


Syntax: [expression for item in iterable if condition]
Purpose: Creates a new list by evaluating an expression for each item in the iterable, optionally including a conditional statement.
When to Use:
When you need more flexibility with the transformation, such as applying conditions (if statements) or when the logic is complex.
When readability and clarity are a priority since list comprehensions are generally more intuitive to read.

Key Differences
Readability: List comprehensions are often more readable and Pythonic, especially for simple transformations.
Performance: The map() function can be faster than a list comprehension in some cases, especially when using built-in functions. However, the performance difference is usually minimal.
Output: map() returns an iterator, so it’s more memory-efficient for large data sets as it doesn’t create a new list in memory until explicitly converted. List comprehensions immediately create a new list.
Function Application: map() is generally more straightforward if you already have a function that needs to be applied, while list comprehensions are better when you need to build a more complex transformation inline.

1. Explain the concept of recursion in Python. How does it differ from iteration?

Recursion and iteration are two fundamental programming concepts used to solve problems, often by repeating a process. They are different in how they handle repetition and problem-solving.

Recursion
Definition: Recursion occurs when a function calls itself directly or indirectly to solve a problem. The idea is to break down a problem into smaller subproblems of the same type, solve the subproblems, and then combine their solutions.

Key Components:

Base Case: A condition that stops the recursion. It's crucial to prevent infinite recursion.
Recursive Case: The part of the function where it calls itself with a modified argument.

```python
def factorial(n):
    # Base case
    if n == 0:
        return 1
    # Recursive case
    else:
        return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

Iteration
Definition: Iteration uses loops to repeat a set of instructions until a condition is met. It does not involve function calls to itself but instead repeatedly executes code blocks until a specific condition is fulfilled.

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))  # Output: 120
```

6. Create a recursive Python function to find the greatest common divisor (GCD) of two numbers using the Euclidean algorithm.

```python
def gcd(a, b):
    # Base case: if b is 0, return a
    if b == 0:
        return a
    # Recursive case: call gcd with b and the remainder of a divided by b
    return gcd(b, a % b)

# Example usage:
num1 = 48
num2 = 18
print(f"The GCD of {num1} and {num2} is {gcd(num1, num2)}")
```

```python
def reverse_string(s):
    # Base case: if the string is empty or a single character
    if len(s) == 0:
        return s
    # Recursive case: reverse the substring and append the first character to the end
    else:
        return reverse_string(s[1:]) + s[0]

# Example usage
input_string = "hello"
reversed_string = reverse_string(input_string)
print(reversed_string)  # Output: "olleh"
```

8. Create a recursive Python function to calculate the power of a number (x^n).

```python
def power(x, n):
    # Base case: x raised to the power of 0 is 1
    if n == 0:
        return 1
    # Base case: x raised to a negative power is 1 / (x raised to the positive power)
    elif n < 0:
        return 1 / power(x, -n)
    # Recursive case: x raised to n is x * (x raised to n-1)
    else:
        return x * power(x, n - 1)

# Example usage:
base = 2
exponent = 5
print(f"{base} raised to the power of {exponent} is {power(base, exponent)}")

# Example with negative exponent:
base = 2
exponent = -3
print(f"{base} raised to the power of {exponent} is {power(base, exponent)}")
```

9. Write a recursive Python function to find all permutations of a given string.

```python
def permutations(s):
    # Base case: if the string is empty, return an empty list
    if len(s) == 0:
        return ['']
    
    # List to store permutations
    perms = []
    
    # Generate permutations for each character
    for i in range(len(s)):
        # Fix the current character
        char = s[i]
        # Generate permutations of the remaining characters
        remaining = s[:i] + s[i+1:]
        for perm in permutations(remaining):
            perms.append(char + perm)
    
    return perms

# Example usage
input_string = "abc"
all_permutations = permutations(input_string)
print(all_permutations)
# Output: ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']
```

10. Write a recursive Python function to check if a string is a palindrome.

```python
def is_palindrome(s):
    # Remove non-alphanumeric characters and convert to lowercase for consistent comparison
    s = ''.join(c.lower() for c in s if c.isalnum())
    
    # Base cases
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    
    # Recursive case
    return is_palindrome(s[1:-1])

# Example usage:
test_string = "A man, a plan, a canal, Panama"
print(f"'{test_string}' is a palindrome: {is_palindrome(test_string)}")

test_string = "Hello, World!"
print(f"'{test_string}' is a palindrome: {is_palindrome(test_string)}")
```

11. Create a recursive Python function to generate all possible combinations of a list of elements.

```python
def combinations(elements):
    # Base case: if the list is empty, return a list with an empty list
    if len(elements) == 0:
        return [[]]
    
    # Recursively generate combinations for the remaining elements
    first = elements[0]
    rest = elements[1:]
    rest_combinations = combinations(rest)
    
    # Add the first element to each combination of the remaining elements
    with_first = [comb + [first] for comb in rest_combinations]
    
    # Combine the results
    return rest_combinations + with_first

# Example usage
input_list = [1, 2, 3]
all_combinations = combinations(input_list)
print(all_combinations)
# Output: [[], [3], [2], [2, 3], [1], [1, 3], [1, 2], [1, 2, 3]]
```

3. Write a Python program that demonstrates the use of global variables within functions.

```python
# Define a global variable
global_variable = 10

def update_global_variable(new_value):
    """Update the global variable with a new value."""
    global global_variable  # Declare that we want to use the global variable
    global_variable = new_value
    print(f"Global variable updated to: {global_variable}")

def use_global_variable():
    """Use the global variable."""
    print(f"Current value of global variable: {global_variable}")

# Main part of the program
print("Initial value of global variable:")
use_global_variable()  # Output: 10

print("\nUpdating global variable...")
update_global_variable(25)  # Updates the global variable to 25

print("\nValue of global variable after update:")
use_global_variable()  # Output: 25
```
